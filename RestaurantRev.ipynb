## Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## a. Top 5 restaurants by revenue

```python
top_revenue = df.nlargest(5, 'Revenue')
print("Top 5 Restaurants by Revenue:")
print(top_revenue[['Name', 'Revenue']])
```

#### The highest earning restaurants generate more revenue than others, indicating potential brand strength or high customer demand
#### The revenue distribution suggests that some top restaurants dominate the market

```python
top_revenue.plot(x='Name', y='Revenue', kind='bar', color='lightblue', edgecolor='royalblue')
plt.title('Top 5 Restaurants by Revenue')
plt.xlabel('Restaurant')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.show()
```

## b. 5 most rated restaurants

```python
top_rated = df.nlargest(5, 'Rating')
print("Top 5 Most Rated Restaurants:")
print(top_rated[['Name', 'Rating']])
```

#### The restaurants with the highest ratings are not the highest earners, suggesting that factors other than customer ratings (eg location, pricing, marketing) impact revenue.
#### High ratings may indicate good service and food quality but do not guarantee high revenue.

```python
top_rated.plot(x='Name', y='Rating', kind='bar', color='pink', edgecolor='purple')
plt.title('Top 5 Most Rated Restaurants')
plt.xlabel('Restaurant')
plt.ylabel('Rating')
plt.xticks(rotation=45)
plt.show()
```
